```python
def N():
  n = 0
  sum = 0
  while sum <= 1:
    sum += random.random()
    n += 1
  return n
```

```python
# a)
def simulacion(nsim):
  sum = 0
  for _ in range(nsim):
    sum += N()
  return sum / nsim
simulacion(1_000_000)
```

```python
# b)
"""
N̄(n) la media muestral de N con n valores de N()
Var(N̄(n)) = s^2 / n
Que aproximamos con S^2 / n

"""
def varianza(nsim):
  n = 1
  media = N()
  s = 0

  while n <= nsim:
    X = N()
    media_ant = media
    media = media_ant + (X - media_ant)/(n+1)
    s = sqrt((1-1/n) * s**2 + (n+1)*(media - media_ant)**2)
    n += 1

  return (s**2)/nsim

varianza(1_000)
```

```python
# c)
def varianza_sim(z_alfa_2, L, nsim):
  d = L / (2 * z_alfa_2)
  s, n = 0, 1
  media = N()
  while n < nsim and (n <= 100 or sqrt(s/n) > d):
    X = N()
    media_ant = media
    media = media_ant + (X - media_ant)/(n+1)
    s = sqrt((1-1/n) * s**2 + (n+1)*(media - media_ant)**2)
    n += 1

  return media

varianza_sim(1.96, 0.025, 10000)
```
