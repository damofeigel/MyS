```python
def binomial(n, p):
  i = 0
  for _ in range(n):
    if random() < p:
      i += 1
  return i

t = 0
for i in range(len(N)):
  t += ((N[i] - n*(1/6))**2)/((1/6)*n)

def sim(T, nsim=10_000):
  pvalor = 0
  frec_muestra = np.zeros(6, int)
  for i in range(nsim):
    frec_muestra *= 0
    for _ in range(n):
      dado = int(random()*6)
      frec_muestra[dado] += 1
    t = 0
    for j in range(6):
      t += (frec_muestra[j] - n*(1/6))**2 / (n*(1/6))
    if t >= T:
      pvalor += 1
  return pvalor/nsim

sim(T=t, nsim=10_000)


```
