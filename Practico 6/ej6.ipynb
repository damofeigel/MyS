```python
from random import random
import numpy as np
```

```python
def random_point():
  u = random()
  v = random()
  if u**2 + v**2 <= 1:
    return 1
  else:
    return 0

def esperanza(d):
  'Estimacion valor esperado con ECM < d'
  media = random_point()
  s, n = 0, 1
  while n <= 100 or np.sqrt(s/n) > d:
    n += 1
    X = random_point()
    media_ant = media
    media = media_ant + (X - media_ant) / (n+1)
    s = s * (1 - 1/n) + (n+1)*(media - media_ant) ** 2

  return 4*media, n, s
```

```python
esperanza(0.001)
```

```python
'''
Para z_alfa_2 = 1.96, L=0.1
si L = 2 * z_alfa_2 * s/sqrt(n)
'''


def simulacion(z_alfa_2, L):
  'Confianza = (1-alfa)%, amplitud del intervalo: L'
  d = L / (2 * z_alfa_2)
  media = random_point()
  s, n = 0, 1
  while n <= 100 or np.sqrt(s/n) > d:
    n += 1
    X = random_point()
    media_ant = media
    media = media_ant + (X - media_ant) / (n+1)
    s = s * (1 - 1/n) + (n+1)*(media - media_ant) ** 2

  return 4*media, n, s
```
