```python
import random, numpy as np
from numpy import sqrt

def fi(x):
  return np.sin(x)/x
def fii(x):
  return 3/(3+x**4)

def integral_ab_semi(fun, a, b, z_alfa_2, nsim, L=0.001):
    n = 1
    media = fun(a+((b-a)*random.random())) * (b-a)
    dev = 0
    semiancho = 0

    while (n <= 100 or L <= semiancho) and n < nsim:
        x = fun(a+((b-a)*random.random())) * (b-a)
        media_ant = media
        media = media_ant + ((x - media_ant)/(n + 1))
        dev = sqrt((1-(1/n))*(dev**2) + (n + 1)*((media - media_ant)**2))
        semiancho = z_alfa_2*dev/sqrt(n)
        n += 1
    return n, media, dev, semiancho*2

def integral_0inf_semi(fun, z_alfa_2, nsim, L=0.001):
    n = 1
    u = random.random()
    media = fun(-1+(1/u)) / (u**2)
    dev = 0
    semiancho = z_alfa_2*dev/sqrt(n)

    while (n <= 100 or L <= semiancho) and n < nsim:
        u = random.random()
        x = fun(-1+(1/u)) / (u**2)
        media_ant = media
        media = media_ant + ((x - media_ant)/(n + 1))
        dev = sqrt((1-(1/n))*(dev**2) + (n + 1)*((media - media_ant)**2))
        semiancho = z_alfa_2*dev/sqrt(n)
        n += 1
    return n, media, dev, semiancho*2
```

```python
def simulacion():
  print(" Nº   |    I    |    S   |    IC")
  for nsim in [1000, 5000, 7000, np.inf]:
    print(integral_ab_semi(fi, np.pi, 2*np.pi, 1.96, nsim))

  print("Ns ↑\n")
  print(" Nº   |    I    |    S   |    IC")
  for nsim in [1000, 5000, 7000, np.inf]:
    print(integral_0inf_semi(fi, 1.96, nsim))
  print("Ns ↑\n")
simulacion()
```
