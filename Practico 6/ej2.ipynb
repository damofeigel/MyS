```python
import random, numpy as np
from numpy import sqrt

def fi(x):
  return np.exp(x) / np.sqrt(2*x)

def fii(x):
  # sqrt(pi) / 2
  return (x**2) * np.exp(-x**2)

def integral_0inf(fun, d=0.01):
  u = random.random()
  media = (1/u**2) * fun(1/u - 1)
  n = 1
  s = 0
  while n <= 100 or sqrt(s/n) <= d:
    n += 1
    u = random.random()
    x = (1/u**2) * fun(1/u - 1)
    media_n = media
    media = media_n + (x - media_n)/(n+1)
    #s = s * (1 - 1/(n-1)) + (n+1)*(media - media_n)**2

  return media, n

def integral_ab(fun, a, b, d=0.01):
  media = fun(a + ((b-a)*random.random())) * (b-a)
  n = 1
  s = 0
  while n <= 100 or sqrt(s/n) >= d:
    n += 1
    u = fun(a + ((b-a)*random.random())) * (b-a)
    media_n = media
    media = media_n + (u - media_n) / (n+1)
    #s = s * (1 - 1/(n-1)) + (n+1)*(media - media_n)**2

  return media, n

def ejercicio2(nsim=10000):
  media_i = 0
  media_ii = 0
  n_i = 0
  n_ii = 0
  for _ in range(nsim):
    sim_i = integral_ab(fi, 0, 1)
    media_i += sim_i[0]
    n_i += sim_i[1]

    sim_ii = integral_0inf(fii)
    media_ii += sim_ii[0]
    n_ii += sim_ii[1]


  print(f"Media 1: {media_i/nsim}")
  print(f"N 1: {n_i/nsim}")
  # Al ser par
  print(f"Media 2: {media_ii*2/nsim}")
  print(f"N 2: {n_ii/nsim}")

ejercicio2()
```
