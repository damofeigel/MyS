```python
import random, numpy as np
from math import sqrt

def simulacion(d=0.1):
  media = np.random.normal(0, 1)
  s = 0
  n = 1
  while n <= 100 or sqrt(s/n) >= d:
    n += 1
    u = np.random.normal(0, 1)
    # Actualizar media
    media_n = media
    media = media_n + (u - media_n) / (n+1)
    # Actualizar Varianza
    s = (1 - 1/n)*s + (n+1)*(media - media_n)**2

  '''
  print(f"Se generaron {generados} datos")
  print(f"La media muestral es {media}")
  print(f"La varianza muestral es {s}")
  '''
  return n, media, sqrt(s)

def ejercicio1(n, nsim=150):
  generados = 0
  media = 0
  s = 0
  for _ in range(nsim):
    tupla = simulacion()
    generados += tupla[0]
    media += tupla[1]
    s += tupla[2]

  print(f"Se generaron {generados/nsim} datos")
  print(f"La media muestral es {media/nsim}")
  print(f"La varianza muestral es {s/nsim}")

ejercicio1(n=150)
```
