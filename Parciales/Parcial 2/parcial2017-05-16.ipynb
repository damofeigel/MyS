```python
p = [0.15, 0.20, 0.10, 0.35, 0.20]

def transformada_inversa(p=p):
  u = random()
  if   u < 0.35:  return 3
  elif u < 0.55:  return 1
  elif u < 0.75:  return 4
  elif u < 0.90:  return 0
  else:           return 2

def Binomial(n,p):
  c = p / (1 - p)
  prob = (1 - p) ** n
  F = prob; i=0
  U = random()
  while U >= F:
    prob *= c * (n-i) / (i+1)
    F += prob
    i += 1
  return i

def Bin(n, p, i):
    return ((factorial(n)/(factorial(n-i)*factorial(i))) * p(i) * (1-p)(n-i))

def aceptacion_rechazo():
  # Esperamos c ciclos
  c = 0.35 / (0.45)**4
  while True:
    u = random()
    y = Binomial(4, 0.45)
    if u < p[y] / (c * Bin(4, 0.45, y)):
      return y
```
