```python
from random import random
from numpy import log, exp

def f(x):
  if x >= 0:
    return exp(-2*x) * (x+1)
  else:
    return 0

def variableY(lamda):
  U = 1-random()
  return -log(U)/lamda

def g(x, lamda):
  if x>= 0:
    return lamda * exp(-lamda*x)
  else:
    return 0

def variableX(lamda=1):
  c = 1
  while True:
    u = random()
    y = variableY(lamda)
    if u < f(y) / (c * g(y)):
      return y
```
